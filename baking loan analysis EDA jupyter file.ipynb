```python
# we are going to ceate a encome band of estimatd encome of clients,because analyzing who are going to give loan back or not
# we have to anaalyze income of peoples who are taking the loan ,so are dividing the encome in band like LOW,MID,HIGH SO WE EASILY analuze encome of people other wise it take time
```

```python
bins = [0,100000,300000,float('inf')]#we divide encom in 0 to 1lac(low),1 to 3lac(mid) and above 3 to infinite (high)
labels = ['LOW','MED','HIGH']# we give the lables of ranges 
df['Income Band'] = pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
# we stored our value of encome labels according to values in new column Income Band
df['Income Band'].value_counts()
# this will no of pepoles whoose income is low and med and high

          
```

```python
numerical_cols =['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
# Univariate analysis and visualiztion
plt.figure(figsize=(15,10))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
    plt.show()
    
    
           
```
